```python
dates = pd.date_range(start="2020-01-01", end="2022-05-16", freq="D")
idx = pd.date_range('01.01.2000', '02.01.2000')

print(df_usersort_user1.shape)
final_df = pd.DataFrame()
for i in range(244):
    
    new_df = df_usersort_user1[df_usersort_user1.iloc[:,2] == list(df_usersort_user1.iloc[:,2].unique())[i]]
    new_df.index = pd.DatetimeIndex(new_df["date"]) 
    new_df = new_df.reindex(dates, fill_value=np.nan)
    new_df.reset_index(inplace=True)
    new_df["index"] = dates
    new_df = new_df.drop(["date"], axis=1)
    new_df.rename(columns= {"index": "date"}, inplace=True)
    new_df['date'] = pd.to_datetime(new_df['date'], format="%Y-%m-%d")
    final_df = pd.concat([final_df, pd.DataFrame.from_records(new_df)])
print(final_df.shape)
final_df.head(10)
print(final_df)
```

```python
final_df_1 = final_df.copy()
final_df.to_csv("Final_DF.csv")
grouped_continent = final_df.groupby(['continent','date'], dropna=False).sum()
grouped_continent["Percentage"] = grouped_continent["total_cases"] / grouped_continent["population"] * 100
grouped_continent["Percentage_vaccines"] = grouped_continent["total_vaccinations"] / grouped_continent["population"] * 100
temp = grouped_continent.index.to_frame()

# deconstruct the date
grouped_continent["year"] = [int(str(temp.iloc[:, 1][i])[0:4]) for i in range(len(temp))]
grouped_continent["month"] = [int(str(temp.iloc[:, 1][i])[5:7]) for i in range(len(temp))]
grouped_continent["day"] = [int(str(temp.iloc[:, 1][i])[8:10]) for i in range(len(temp))]
grouped_continent.to_csv("Continent_grouped.csv")

grouped_continent
```

```python
final_df_1 = final_df.copy()
iso_codes = pd.read_csv("Iso_codes.csv")
iso_codes = iso_codes.rename(columns={"English short name lower case" : "location"})
iso_codes = iso_codes.rename(columns={"Alpha-3 code" : "iso_code"})
grouped_countries = final_df.groupby(['continent', 'location','date'], dropna=False).sum()
grouped_countries["Percentage"] = grouped_countries["total_cases"] / grouped_countries["population"] * 100
grouped_countries["Percentage_vaccines"] = grouped_countries["total_vaccinations"] / grouped_countries["population"] * 100
temp = grouped_countries.index.to_frame()

# deconstruct the date
grouped_countries["year"] = [int(str(temp.iloc[:, 2][i])[0:4]) for i in range(len(temp))]
grouped_countries["month"] = [int(str(temp.iloc[:, 2][i])[5:7]) for i in range(len(temp))]
grouped_countries["day"] = [int(str(temp.iloc[:, 2][i])[8:10]) for i in range(len(temp))]
grouped_countries["continent"] = [x[0] for x in grouped_countries.index]
grouped_countries["complete_dates"] = [x[2] for x in grouped_countries.index]
grouped_countries = grouped_countries.merge(iso_codes, "left", on="location")
grouped_countries.to_csv("countries_grouped.csv")

grouped_countries
```
